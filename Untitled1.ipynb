```
#detect full face
def detect_face_FUll(img):
 face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
 #cv2_imshow(img)
 faces = face_cascade.detectMultiScale(img, 1.1, 4)
 for (x, y, w, h) in faces:
   img = img[y:y+h, x:x+w]
 #cv2_imshow( img)
 img =cv2.resize(img, (39,39))#(width,height)
 #cv2_imshow( img)
 return img

#detect right face
def detect_face_righthalf(img):
 face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
 #cv2_imshow(img)
 faces = face_cascade.detectMultiScale(img, 1.1, 4)
 for (x, y, w, h) in faces:
    img = img[y:y+h, x:x+w]
 #cv2_imshow( img)
 img =cv2.resize(img, (39,39))#(width,height)
 #cv2_imshow( img)
 img = img[0:39, 12:39]
 return img

#detect left face
def detect_face_lefthalf(img):
 face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
 #cv2_imshow(img)
 faces = face_cascade.detectMultiScale(img, 1.1, 4)
 for (x, y, w, h) in faces:
   img = img[y:y+h, x:x+w]
 #cv2_imshow( img)
 img =cv2.resize(img, (39,39))#(width,height)
 #cv2_imshow( img)
 img = img[0:39, 0:27]
 return img
```

```

model = Sequential()
model.add(Conv2D(20, (4, 4),input_shape=(39,39,1),activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(20, (3, 3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(20, (3, 3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(20, (2, 2),activation="relu"))

model.add(Flatten())

model.add(Dense(80,activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(4,activation="relu"))

model.summary()

opt = keras.optimizers.sgd(learning_rate=0.0001, decay=1e-6)
model.compile(loss='mean_squared_error',optimizer=opt,metrics=['accuracy'])

model.fit(Xtrain, Ytrain, batch_size=500, epochs=10, validation_data = (Xtest, Ytest), verbose = 1)
model.save('F1.h5')
```
